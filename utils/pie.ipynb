```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def read_list_and_draw_pie(input_list:list):
	name_list = []
	num_list = []
	for item in input_list:
		name_list.append(item[0])
		num_list.append(item[1])
	print(num_list)
	print(name_list)
	# 保证圆形
	plt.axes(aspect=1)
	plt.pie(x=num_list, labels=name_list, autopct='%3.1f %%')
	plt.show()

read_list_and_draw_pie([('A', 2),  ('B',3), ('C', 5)])
```

```python
def read_excel_and_draw_pie(filename):

	df = pd.read_csv(filename, encoding='utf-8')
	# print(df.head())
	a = df.groupby('分类').size()
	print('statistic of domain data', df.groupby('分类').size())
	'''
	visualization
	'''
	a.plot.pie(figsize=(20, 20), autopct='%.2f')
	# http://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html
	plt.show()

read_excel_and_draw_pie('/Users/sunhongchao/Documents/craft/Workspace-of-NLU/corpus/clf/jingling.csv')
```
