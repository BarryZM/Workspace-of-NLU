# Confidence Interval

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

N = 10000
x = np.random.normal(0, 1, N)
# ddof取值为1是因为在统计学中样本的标准偏差除的是(N-1)而不是N，统计学中的标准偏差除的是N
# SciPy中的std计算默认是采用统计学中标准差的计算方式
mean, std = x.mean(), x.std(ddof=1)
print(mean, std)
# 计算置信区间
# 这里的0.9是置信水平
conf_intveral = stats.norm.interval(0.9, loc=mean, scale=std)
print(conf_intveral)

conf_intveral = stats.norm.interval(0.5, loc=mean, scale=std)
print(conf_intveral)

conf_intveral = stats.norm.interval(0.1, loc=mean, scale=std)
print(conf_intveral)

# 绘制概率密度分布图
x = np.arange(-5, 5, 0.001)
# PDF是概率密度函数
y = stats.norm.pdf(x, loc=mean, scale=std)
plt.plot(x, y)
plt.show()
```

```python
# 随机数
N = 300
x = np.random.rand(N)
mean, std = x.mean(), x.std(ddof=1)
print(mean, std)

conf_intveral = stats.norm.interval(0.5, loc=mean, scale=std)
print(conf_intveral)

# 绘制概率密度分布图
x = np.arange(-5, 5, 0.001)
# PDF是概率密度函数
y = stats.norm.pdf(x, loc=mean, scale=std)
plt.plot(x, y)
plt.show()
```

```python
# 设定 多分类ROC的置信区间
```

```python
# 输入， 模型
# 一批数据， 判断置信区间

import numpy
from sklearn.metrics import accuracy_score

outptus = numpy.random.randint(low=0, high=10, size = 100) 
outptus_1 = outptus.copy()
grounds = [item if item != 3 else 9 for item in outptus_1]

acc = accuracy_score(grounds, outptus_1)
print(acc)

from math import sqrt
interval = 4 * sqrt( (acc * (1 - acc)) / 300)
print('%.3f' % interval)

```
